```python
# 3.5 Finding Median
def find_median(x):
    if len(x) % 2 == 0:
        a = sorted(x)
        median = (a[len(x)//2] + a[len(x)//2 -1])/2
        return median
    else:
        a = sorted(x)
        median = (a[len(x)//2])
        return median
find_median([27,48,9,63,99,61,33,80,43,84,39,46,40,46,16,55,69,43,11,57]) 
```

```python
# 3.6 middle point
def middle_point(x,y):
    a,b = x 
    c,d = y
    m1 = (a+c)/2
    m2 = (b+d)/2
    return (m1,m2)

middle_point((-4.0, 2.0), (3.0, -4.0))
```

```python
# 3.7 Sampled Average
def sampled_average(numbers,k):
    n = len(numbers)
    if sum(numbers) - k * (n-1) in numbers:
        return "yes"
    else:
        return "no"
    
```

```python
# 3.9 Bankruptcy Countdown
def countdown(x,y):
    monthly_count = 0
    for i in y:
        if x >= i:
            monthly_count += 1
            x -= i
        elif x < i:
            break
    return monthly_count
countdown(12500, [7000, 4500, 750, 150, 7100, 1000])
countdown(2000, [2900, 1000, 400, 23000])
```

```python
# 3.12 Palindrome 
def is_palindrome(x):
    a = []
    for i in x:
        a.append(i)
    b = []
    for i in a[::-1]:
        b.append(i)
    if a == b:
        return True
    else:
        return False
    
is_palindrome('aba')
```

```python
# 3.13 Range of 
def range_of(x):
    a = x[0] # min
    b = x[0] # max
    for i in x:
        if i > b:
            b = i
        elif i < a:
            a = i
    return b - a
range_of([1])
range_of([1, 4, 6, 28, -10, 0, 0, 1])
```

```python
## 3.16 Substring Count
def substring_count(x,y):
    count = 0
    end = len(y) 
    for i in range(0,len(x)):
        if x[i:end] == y :
            count += 1
        end += 1
    return count

substring_count("ababa","aba")
substring_count("stanrandystancartmanstankenny","stan")
        
```

```python
# 3.17 reduce number
def sum(x):
    total = 0
    for i in str(x):
        total += int(i)
    return total

def reduce_number(y):
    while y >= 10:
        y = sum(y)
    return y
     
print(reduce_number(589564123))
print(reduce_number(19999999999999))        

```

```python
# 3.18 my range
def my_range(start,stop,step):
    a = []
    current = start
    if step > 0:
        while current <stop:
            a.append(current)
            current += step
    elif step <0:
        while current >stop:
            a.append(current)
            current += step
    return a
my_range(3, 10, 1)
```

```python
## 3.19 Even Numbers without % operator
def is_even(x):
    if x == 1:
        return False
    if x == 0 and x == 2:
        return True
    if x < 0:
        x = abs(x)
    while x > 2:
        x -= 2
    if x == 1:
        return False
    else:
        return True
is_even(0)
is_even(-15)
```

```python
## 3.20 First Non-Prime
## first we are starting with prime_number function
def is_prime(x):
    if x == 1:
        return False
    if x == 2:
        return True
    count = 0
    if x == 3:
        return True
    if x == 4:
        return False
    for i in range(3,x):
        if x % i == 0: 
            count += 1
        if count == 0:
            return True
        else:
            return False
## test it
is_prime(3)
is_prime(1097) ## 1097 is a prime number

def find_first_non_prime(x):
    a = []
    for i in x:
        if is_prime(i) == False:
            a.append(i)
    if len(a) == 0:
        return False
    else:
        return a[0]
        
find_first_non_prime([2,3,4,7])
find_first_non_prime([2,3,51,19]) 

## not perfect :)))
```

```python
### 3.21 The Game Of Divisors
def divisor_number(k):
    a =[]
    for i in range(1,k):
        if k % i == 0:
            a.append(i)
    return len(a)

def winner(nazif,osman):
    if divisor_number(nazif) > divisor_number(osman):
        return "Nazif"
    elif divisor_number(nazif) < divisor_number(osman):
        return "Osman"
    else:
        return "Draw"
winner(4,121)
winner(120,144)
winner(191,190)
```

```python
## 3.22 Perfect Number
def is_perfect(k):
    a =[]
    for i in range(1,k):
        if k % i == 0:
            a.append(i)
    total = 0
    for i in a:
        total += i
    if total == k:
        return True
    else:
        return False
    
print(is_perfect(6))
print(is_perfect(7))
print(is_perfect(28))
print(is_perfect(30))
```

```python
## 3.23 Anagram
def letter_list_find(l):
    letter_list = []
    for i in l:
        if i in letter_list:
            continue
        else:
            letter_list.append(i)
    return sorted(letter_list)
def anagram(x,y):
    if letter_list_find(x) == letter_list_find(y):
        return True
    else:
        return False
## instead of using dictionaries firstly i create a list that 
## contains every letter(just once) without order, then it
## returned a sorted list to make every same letter list with
## the same order.
print(anagram('palm', 'lamp'))
print(anagram('hello', 'olla'))
```
